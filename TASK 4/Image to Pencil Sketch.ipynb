## Task 4: Image to Pencil Sketch with Python

**The objective of this project is to create a pencil sketch of normal image file**




## Importing Required Python Libraries

```python
import cv2
from matplotlib import pyplot as plt
import matplotlib.image as mpimg
```

```python
img = cv2.imread('myimage.jpg')
```

```python
plt.imshow(img)
b,q,r = cv2.split(img)
img = cv2.merge((r,q,b))
```

## Original image looks like this....

```python
plt.figure(figsize=(8,8))
plt.title("Original Image", fontsize=20)
plt.imshow(img)
```

## Converting original image to grayscale image.

```python
img = mpimg.imread('myimage.jpg')
R, G, B = img[:,:,0], img[:,:,1], img[:,:,2]
imgGray = 0.2989 * R + 0.5870 * G + 0.1140 * B
plt.figure(figsize=(8,8))
plt.imshow(imgGray, cmap='gray')
plt.title("GrayScale Image", fontsize=20)
plt.show()
```

## Now, grayscale image is inverted...

```python
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
img_invert = cv2.bitwise_not(img_gray)
plt.figure(figsize=(8,8))
plt.imshow(img_invert,cmap="gray")
plt.axis("off")
plt.title("Inverted Image", fontsize=20)
plt.show()
```

## It's time to smoothen the inverted image.

```python
img_smoothing = cv2.GaussianBlur(img_invert, (21, 21),sigmaX=0, sigmaY=0)
plt.figure(figsize=(8,8))
plt.imshow(img_smoothing,cmap="gray")
plt.axis("off")
plt.title("Smoothen Image", fontsize=20)
plt.show()
```

## The final pencil sketch generated for the original image is:

```python
final = cv2.divide(img_gray, 255 - img_smoothing, scale=255)
plt.figure(figsize=(8,8))
plt.imshow(final,cmap="gray")
plt.axis("off")
plt.title("Final Sketch Image", fontsize=20)
plt.show()
```

## So, a quick comaprision between original image and pencil sketch is:

```python
plt.figure(figsize=(20,20))
plt.subplot(1,2,1)
plt.title('Original image',  size=20)
plt.imshow(img)
plt.axis('off')
plt.subplot(1,2,2)
plt.title('Final Pencil sketch', size=20)
final=cv2.cvtColor( final , cv2.COLOR_BGR2RGB)
plt.imshow( final )
plt.axis('off')
plt.show()
```
